```python
sql(""" SELECT
    tb.tconst AS tconst,
    tb.primaryTitle AS Title,
    tb.genres AS Genres,
    tb.startYear AS Year,
    tb.runtimeMinutes AS Runtime,
    string_agg(DISTINCT actors.primaryName, ', ') AS Actors,
    directors.primaryName AS Director,
    t.overview AS Resume,
    tr.averageRating AS Rating,
    tr.numVotes AS Votes,
    t.Budget AS Budget,
    t.Revenue AS Revenue,
    t.production_companies_name AS ProductionCompanies,
    t.production_companies_country AS ProductionCountries,
    t.poster_path AS Poster
FROM title_basics tb
JOIN title_akas ta ON tb.tconst = ta.titleId
JOIN title_crew tc ON tb.tconst = tc.tconst
JOIN name_basics directors ON tc.directors = directors.nconst
JOIN title_principals tp ON tb.tconst = tp.tconst
JOIN name_basics actors ON tp.nconst = actors.nconst
JOIN tmdb t ON tb.tconst = t.imdb_id
JOIN title_ratings tr ON tb.tconst = tr.tconst
WHERE tp.category IN ('actor', 'actress')
GROUP BY tb.tconst, tb.primaryTitle, tb.genres, tb.startYear, tb.runtimeMinutes,
         directors.primaryName, t.overview, tr.averageRating, tr.numVotes,
         t.Budget, t.Revenue, t.production_companies_name,
         t.production_companies_country, t.poster_path
ORDER BY t.Revenue DESC
LIMIT 1
""")
```

```python
sql("""
WITH writers_expanded AS (
    SELECT
        tc.tconst,
        nb.primaryName AS Writer
    FROM title_crew tc
    CROSS JOIN UNNEST(string_split(tc.writers, ',')) AS writer_nconst(nconst)
    JOIN name_basics nb ON nb.nconst = writer_nconst.nconst
)

SELECT
    tb.tconst AS tconst,
    tb.primaryTitle AS Title,
    tb.genres AS Genres,
    tb.startYear AS Year,
    tb.runtimeMinutes AS Runtime,
    string_agg(DISTINCT actors.primaryName, ', ') AS Actors,
    string_agg(DISTINCT directors.primaryName, ', ') AS Director,
    string_agg(DISTINCT we.Writer, ', ') AS Writers,
    t.overview AS Resume,
    tr.averageRating AS Rating,
    tr.numVotes AS Votes,
    t.Budget AS Budget,
    t.Revenue AS Revenue,
    t.Revenue - t.Budget AS Profit,
    t.production_companies_name AS ProductionCompanies,
    t.production_companies_country AS ProductionCountries,
    t.poster_path AS Poster
FROM title_basics tb
JOIN title_akas ta ON tb.tconst = ta.titleId
JOIN title_crew tc ON tb.tconst = tc.tconst
CROSS JOIN UNNEST(string_split(tc.directors, ',')) AS director_nconst(nconst)
JOIN name_basics directors ON directors.nconst = director_nconst.nconst
JOIN title_principals tp ON tb.tconst = tp.tconst
JOIN name_basics actors ON tp.nconst = actors.nconst
JOIN tmdb t ON tb.tconst = t.imdb_id
JOIN title_ratings tr ON tb.tconst = tr.tconst
LEFT JOIN writers_expanded we ON tb.tconst = we.tconst
WHERE tp.category IN ('actor', 'actress')
    and Runtime between 60 and 240
    and tb.startYear > '1940'
    
    and Votes > 100
GROUP BY tb.tconst, tb.primaryTitle, tb.genres, tb.startYear, tb.runtimeMinutes,
         t.overview, tr.averageRating, tr.numVotes,
         t.Budget, t.Revenue, t.production_companies_name,
         t.production_companies_country, t.poster_path
ORDER BY Profit DESC

""")
```

```python
sql(""" CREATE OR REPLACE VIEW dataframe_view AS 
    
WITH writers_expanded AS (
    SELECT
        tc.tconst,
        nb.primaryName AS Writer
    FROM title_crew tc
    CROSS JOIN UNNEST(string_split(tc.writers, ',')) AS writer_nconst(nconst)
    JOIN name_basics nb ON nb.nconst = writer_nconst.nconst
)

SELECT
    tb.tconst AS tconst,
    tb.primaryTitle AS Title,
    tb.genres AS Genres,
    tb.startYear AS Year,
    tb.runtimeMinutes AS Runtime,
    string_agg(DISTINCT actors.primaryName, ', ') AS Actors,
    string_agg(DISTINCT directors.primaryName, ', ') AS Director,
    string_agg(DISTINCT we.Writer, ', ') AS Writers,
    t.overview AS Resume,
    tr.averageRating AS Rating,
    tr.numVotes AS Votes,
    t.Budget AS Budget,
    t.Revenue AS Revenue,
    t.Revenue - t.Budget AS Profit,
    t.production_companies_name AS ProductionCompanies,
    t.production_companies_country AS ProductionCountries,
    t.poster_path AS Poster
FROM title_basics tb
JOIN title_akas ta ON tb.tconst = ta.titleId
JOIN title_crew tc ON tb.tconst = tc.tconst
CROSS JOIN UNNEST(string_split(tc.directors, ',')) AS director_nconst(nconst)
JOIN name_basics directors ON directors.nconst = director_nconst.nconst
JOIN title_principals tp ON tb.tconst = tp.tconst
JOIN name_basics actors ON tp.nconst = actors.nconst
JOIN tmdb t ON tb.tconst = t.imdb_id
JOIN title_ratings tr ON tb.tconst = tr.tconst
LEFT JOIN writers_expanded we ON tb.tconst = we.tconst
WHERE tp.category IN ('actor', 'actress')
    and Runtime between 60 and 240
    and tb.startYear > '1940'
    
    and Votes > 100
GROUP BY tb.tconst, tb.primaryTitle, tb.genres, tb.startYear, tb.runtimeMinutes,
         t.overview, tr.averageRating, tr.numVotes,
         t.Budget, t.Revenue, t.production_companies_name,
         t.production_companies_country, t.poster_path
ORDER BY Profit DESC



""")
```
